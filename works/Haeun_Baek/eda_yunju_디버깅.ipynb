```python
import pandas as pd

other_files2 = [
    './tft_game_dataset/TFT_Challenger_MatchData.csv',
    './tft_game_dataset/TFT_GrandMaster_MatchData.csv',
    './tft_game_dataset/TFT_Master_MatchData.csv',
    './tft_game_dataset/TFT_Diamond_MatchData.csv',
    './tft_game_dataset/TFT_Platinum_MatchData.csv'
]

summary = []

for file in other_files2:
    df = pd.read_csv(file)
    # Set3_ 라벨이 포함된 행의 수 계산
    s3_count = df['combination'].str.contains('Set3_', na=False).sum()
    total = len(df)
    ratio = (s3_count / total) * 100
    
    summary.append({
        '티어': file.split('_')[1],
        '전체 데이터': total,
        '시즌3 확정(Set3_)': s3_count,
        '오염된 데이터': total - s3_count,
        '정상 비율(%)': round(ratio, 2)
    })

df_summary = pd.DataFrame(summary)
print("[전 티어 데이터 무결성 검사 결과]")
display(df_summary)
```

```python
dataframes = {
    'Challenger': df_Challenger,
    'GrandMaster': df_GrandMaster,
    'Master': df_Master,
    'Diamond': df_Diamond,
    'Platinum': df_Platinum,
}

final_report = []

print("[전 티어 데이터 보존율 검사 시작]")
print("-" * 60)

for tier_name, df in dataframes.items():
    total_count = len(df)
    
    s3_labeled = df[df['combination'].str.contains('Set3_', na=False)]
    s3_count = len(s3_labeled)
    survival_rate = (s3_count / total_count) * 100
    
    final_report.append({
        'Tier': tier_name,
        'Original': total_count,
        'Survived(Set3_)': s3_count,
        'Survival Rate(%)': round(survival_rate, 2)
    })
    
    print(f"{tier_name:12} | 보존량: {s3_count:,} / {total_count:,} ({survival_rate:.2f}%)")

print("-" * 60)
df_report = pd.DataFrame(final_report)
```

```python
import pandas as pd

dfs = {
    'Challenger': df_Challenger,
    'GrandMaster': df_GrandMaster,
    'Master': df_Master,
    'Diamond': df_Diamond,
    'Platinum': df_Platinum,
}

summary_list = []

print("[티어별 시즌 데이터 비중 분석]")
print("-" * 60)

for name, df in dfs.items():
    total_len = len(df)
    
    has_s2 = df['combination'].str.contains('Set2_', case=False, na=False)
    has_s3 = df['combination'].str.contains('Set3_', case=False, na=False)
    
    s3_count = (has_s3 & ~has_s2).sum()  # 순수 s3만
    ratio = (s3_count / total_len) * 100

    summary_list.append({
        '티어(Tier)': name,
        '전체 행 수': f"{total_len:,}",
        's2_only': f"{(has_s2 & ~has_s3).sum():,}",
        's3_only': f"{s3_count:,}",
        'mixed': f"{(has_s2 & has_s3).sum():,}",
        'neither': f"{(~has_s2 & ~has_s3).sum():,}",
        '시즌 3 비중': f"{ratio:.2f}%"
    })

summary_df = pd.DataFrame(summary_list)
display(summary_df)
print("-" * 60)
```

```python
import ast

season2_traits = {
    "Steel", "Shadow", "Lunar", "Mountain", "Poison", "Ocean", "Wind", "Set2_Glacial",
    "Light", "Desert", "Woodland", "Electric", "Inferno", "Set2_Blademaster", "Berserker",
    "Druid", "Summoner", "Mystic", "Avatar", "Set2_Assassin", "Alchemist", "Soulbound",
    "Mage", "Set2_Ranger", "Warden", "Predator", "Crystal"
}

def has_season2_trait(trait_list):
    return bool(set(trait_list) & season2_traits)

summary_list = []

print("[티어별 시즌 데이터 비중 분석]")
print("-" * 60)

for name, df in dfs.items():
    df['combination'] = df['combination'].apply(ast.literal_eval)
    
    has_s2 = df['combination'].apply(has_season2_trait)
    has_s3 = ~has_s2  # S2 목록에 없으면 S3
    
    total_len = len(df)
    s3_count = has_s3.sum()
    ratio = (s3_count / total_len) * 100

    summary_list.append({
        '티어(Tier)': name,
        '전체 행 수': f"{total_len:,}",
        's2_only': f"{has_s2.sum():,}",
        's3_only': f"{s3_count:,}",
        '시즌 3 비중': f"{ratio:.2f}%"
    })

summary_df = pd.DataFrame(summary_list)
display(summary_df)
print("-" * 60)
```
