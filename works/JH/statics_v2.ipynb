### 1번: 특정 시너지 조합은 다른 조합보다 유의미하게 승률이 높다

```python
import pandas as pd
from scipy.stats import chi2_contingency

df_synergy = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_combination_1.csv')
results = []

for syn in df_synergy['synergy'].unique():
    df_synergy['is_target'] = (df_synergy['synergy'] == syn)
    contingency = pd.crosstab(df_synergy['is_target'], df_synergy['ranked_1'])
    
    if contingency.shape == (2, 2):
        chi2, p, dof, expected = chi2_contingency(contingency)
        target_win_rate = df_synergy[df_synergy['is_target'] == True]['ranked_1'].mean()
        other_win_rate = df_synergy[df_synergy['is_target'] == False]['ranked_1'].mean()
        
        results.append({
            'synergy': syn,
            'target_win_rate': target_win_rate,
            'other_win_rate': other_win_rate,
            'p_value': p
        })

result_df = pd.DataFrame(results).sort_values('p_value')
display(result_df)
```

결과 표에서 p_value가 0.05 미만이면서 target_win_rate가 other_win_rate보다 눈에 띄게 높은 시너지들이 바로 우리가 찾는 다른 조합보다 유의미하게 승률이 높은 시너지

### 11번: 특정 아이템 조합이 챔피언 성과에 유의미한 영향을 준다

```python
import pandas as pd
from scipy.stats import ttest_ind

df_item = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_items_1.csv')
results = []

for champ in df_item['champion_name'].unique():
    champ_data = df_item[df_item['champion_name'] == champ]
    
    for item_comb in champ_data['items'].unique():
        target_group = champ_data[champ_data['items'] == item_comb]['ranked']
        other_group = champ_data[champ_data['items'] != item_comb]['ranked']
        
        if len(target_group) > 1 and len(other_group) > 1:
            stat, p = ttest_ind(target_group, other_group, equal_var=False)
            
            results.append({
                'champion': champ,
                'item_combination': item_comb,
                'target_mean_rank': target_group.mean(),
                'other_mean_rank': other_group.mean(),
                'p_value': p
            })

result_df = pd.DataFrame(results).sort_values('p_value')
display(result_df)
```

출력된 데이터프레임에서 p_value가 0.05 미만이면서 target_mean_rank가 other_mean_rank보다 더 낮은(TFT에서 등수는 숫자가 작을수록 좋으므로) 항목들이 바로 그 챔피언에게 다른 템트리보다 통계적으로 확실히 더 좋은 성과를 내게 해주는 특정 아이템 조합

### 13번: 특정 시너지는 1위 달성률이 전체 평균보다 유의미하게 높다

```python
import pandas as pd
from scipy.stats import binomtest

df_synergy = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_combination_1.csv')
overall_win_rate = df_synergy['ranked_1'].mean()
results = []

for syn in df_synergy['synergy'].unique():
    target_data = df_synergy[df_synergy['synergy'] == syn]
    successes = target_data['ranked_1'].sum()
    trials = len(target_data)
    
    if trials > 0:
        res = binomtest(successes, n=trials, p=overall_win_rate, alternative='greater')
        
        results.append({
            'synergy': syn,
            'target_win_rate': successes / trials,
            'overall_win_rate': overall_win_rate,
            'p_value': res.pvalue
        })

result_df = pd.DataFrame(results).sort_values('p_value')
display(result_df)
```

출력된 표에서 p_value가 0.05 미만이면서 target_win_rate가 overall_win_rate보다 높은 시너지들이 바로 전체 평균을 웃도는 진짜 1위 달성률 1티어 시너지들 일것인데 p벨류가 아까부터 거 참;;

### 2번: 픽률이 높은 시너지가 반드시 높은 승률을 보인다.

```python
import pandas as pd
from scipy.stats import spearmanr

df_synergy = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_combination_1.csv')

synergy_stats = df_synergy.groupby('synergy').agg(
    pick_count=('synergy', 'count'),
    win_rate=('ranked_1', 'mean')
).reset_index()

stat, p_value = spearmanr(synergy_stats['pick_count'], synergy_stats['win_rate'])

print(f"Spearman correlation: {stat}, p-value: {p_value}")
```

## 8번: 시너지별 픽률에 유의미한 차이가 있다.

```python
import pandas as pd
from scipy.stats import binomtest

df_synergy = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_combination_1.csv')

total_boards = len(df_synergy[['gameid', 'user_id']].drop_duplicates())
synergy_counts = df_synergy['synergy'].value_counts()
mean_pick_rate = (synergy_counts / total_boards).mean()

results = []
for syn, count in synergy_counts.items():
    res = binomtest(count, n=total_boards, p=mean_pick_rate, alternative='two-sided')
    
    results.append({
        'synergy': syn,
        'pick_rate': count / total_boards,
        'mean_pick_rate': mean_pick_rate,
        'p_value': res.pvalue
    })

result_df = pd.DataFrame(results).sort_values('p_value')
display(result_df)
```

출력된 표에서 p_value가 0.05 미만인 시너지들이 유의미한 픽률 차이를 보이는 항목들, pick_rate가 mean_pick_rate보다 높으면 유저들이 비정상적으로 많이 찾는 오버파워(OP) 혹은 인기 시너지라는 뜻, 낮으면 철저히 소외된 비주류 시너지라는 뜻으로 해석 가능!

## 9번: 픽률 상위 시너지가 승률도 높은가.

```python
import pandas as pd
from scipy.stats import linregress

df_synergy = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_combination_1.csv')

synergy_stats = df_synergy.groupby('synergy').agg(
    pick_count=('synergy', 'count'),
    win_rate=('ranked_1', 'mean')
).reset_index()

res = linregress(synergy_stats['pick_count'], synergy_stats['win_rate'])

print("Slope:", res.slope)
print("R-squared:", res.rvalue**2)
print("p-value:", res.pvalue)
```

출력된 p-value가 0.05 미만이면 픽률이 승률에 유의미한 영향을 준다는 뜻
Slope(기울기) 값이 양수(+)이면 픽률이 오를수록 승률도 올라간다는 뜻이고, 음수(-)이면 픽률이 오를수록 승률은 떨어진다는 뜻으로 해석
R-squared 값은 1에 가까울수록 픽률이 승률을 아주 잘 설명한다는 뜻

## 12번: 픽률 하위임에도 승률 상위인 시너지가 통계적으로 유의미하게 존재한다. (꿀덱 찾기)

```python
import pandas as pd
from scipy.stats import binomtest

df_synergy = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_combination_1.csv')

overall_win_rate = df_synergy['ranked_1'].mean()
synergy_counts = df_synergy['synergy'].value_counts()
low_pick_threshold = synergy_counts.quantile(0.25)

low_pick_synergies = synergy_counts[synergy_counts <= low_pick_threshold].index
results = []

for syn in low_pick_synergies:
    target_data = df_synergy[df_synergy['synergy'] == syn]
    successes = target_data['ranked_1'].sum()
    trials = len(target_data)
    
    if trials > 0:
        res = binomtest(successes, n=trials, p=overall_win_rate, alternative='greater')
        results.append({
            'synergy': syn,
            'pick_count': trials,
            'win_rate': successes / trials,
            'p_value': res.pvalue
        })

result_df = pd.DataFrame(results).sort_values('p_value')
display(result_df)
```

### 17번: 티어와 관계없이 언제나 성과가 좋은 국밥 챔피언 찾기

```python
import pandas as pd
from scipy.stats import f_oneway

df_champ = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_champion_1.csv')
results = []

for champ in df_champ['champion_name'].unique():
    champ_data = df_champ[df_champ['champion_name'] == champ]
    tier_groups = [group['ranked'].values for name, group in champ_data.groupby('tier') if len(group) > 5]
    
    if len(tier_groups) > 1:
        stat, p = f_oneway(*tier_groups)
        results.append({
            'champion': champ,
            'mean_rank': champ_data['ranked'].mean(),
            'p_value': p
        })

result_df = pd.DataFrame(results).sort_values('mean_rank')
display(result_df)
```

### 문제 5-1: 티어별 3성 효율 차이

```python
import pandas as pd
from scipy.stats import ttest_ind

df_champ = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_champion_1.csv')

player_stats = df_champ.groupby(['gameid', 'user_id', 'tier']).agg(
    max_star=('star', 'max'),
    ranked=('ranked', 'first')
).reset_index()

tier_order = ['platinum', 'diamond', 'master', 'grand_master', 'challenger']
results = []

for t in tier_order:
    tier_data = player_stats[player_stats['tier'] == t]
    
    has_3_star = tier_data[tier_data['max_star'] >= 3]['ranked']
    no_3_star = tier_data[tier_data['max_star'] < 3]['ranked']
    
    if len(has_3_star) > 1 and len(no_3_star) > 1:
        stat, p_value = ttest_ind(has_3_star, no_3_star, equal_var=False)
        
        results.append({
            'tier': t,
            'has_3_star_mean_rank': has_3_star.mean(),
            'no_3_star_mean_rank': no_3_star.mean(),
            'rank_diff': no_3_star.mean() - has_3_star.mean(),
            'p_value': p_value
        })

result_df = pd.DataFrame(results)
result_df['tier'] = pd.Categorical(result_df['tier'], categories=tier_order, ordered=True)
result_df = result_df.sort_values('tier')
display(result_df)
```

### 5-2번: 챔피언 코스트별 3성 가치 차이

```python
import pandas as pd
from scipy.stats import f_oneway

df_champ = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_champion_1.csv')

# 3성 챔피언 데이터만 추출
df_3_star = df_champ[df_champ['star'] == 3].copy()

# 코스트별 등수 데이터를 담을 리스트
cost_groups = []
results = []

# 코스트(1~5)별로 순회
for cost in sorted(df_3_star['cost'].dropna().unique()):
    ranks = df_3_star[df_3_star['cost'] == cost]['ranked']
    
    if len(ranks) > 0:
        cost_groups.append(ranks.values)
        results.append({
            'cost': cost,
            '3_star_count': len(ranks),
            'mean_rank': ranks.mean()
        })

# 코스트 간 등수 차이가 유의미한지 분산분석(ANOVA) 진행
if len(cost_groups) > 1:
    stat, p_value = f_oneway(*cost_groups)
else:
    p_value = None

print(f"ANOVA p-value: {p_value}\n")

result_df = pd.DataFrame(results)
display(result_df)
```

### 5-3번: 3성 기물의 개수와 승률의 관계

### 5-4 3성작이 강제되는 챔피언이 있을까?

```python
import pandas as pd
from scipy.stats import ttest_ind

df_champ = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_champion_1.csv')
results = []

for champ in df_champ['champion_name'].dropna().unique():
    champ_data = df_champ[df_champ['champion_name'] == champ]
    
    star2 = champ_data[champ_data['star'] == 2]['ranked']
    star3 = champ_data[champ_data['star'] == 3]['ranked']
    
    if len(star2) > 5 and len(star3) > 5:
        stat, p_value = ttest_ind(star2, star3, equal_var=False)
        
        results.append({
            'champion': champ,
            '2_star_mean_rank': star2.mean(),
            '3_star_mean_rank': star3.mean(),
            'rank_diff': star2.mean() - star3.mean(),
            'p_value': p_value
        })

result_df = pd.DataFrame(results).sort_values('rank_diff', ascending=False)
display(result_df)
```

### 7번: 같은 코스트 내에서도 챔피언 간에 유의미한 평균 등수 차이가 존재한다

```python
import pandas as pd
from scipy.stats import kruskal

df_champ = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_champion_1.csv')
df_valid = df_champ.dropna(subset=['cost', 'champion_name', 'ranked'])

results = []

# 코스트별로 순회 (1 ~ 5 코스트)
for cost in sorted(df_valid['cost'].unique()):
    cost_data = df_valid[df_valid['cost'] == cost]
    
    champ_groups = []
    champ_names = []
    
    # 해당 코스트 내의 챔피언별로 데이터 그룹화
    for champ, group in cost_data.groupby('champion_name'):
        if len(group) > 30: # 표본이 너무 적은 챔피언은 제외 (신뢰도 확보)
            champ_groups.append(group['ranked'].values)
            champ_names.append(champ)
    
    # 비교할 챔피언이 2명 이상일 때만 검정 수행
    if len(champ_groups) > 1:
        stat, p_value = kruskal(*champ_groups)
        
        # 챔피언별 평균 등수 계산
        mean_ranks = cost_data.groupby('champion_name')['ranked'].mean().reset_index()
        mean_ranks = mean_ranks[mean_ranks['champion_name'].isin(champ_names)]
        
        # 가장 좋은 챔피언 (등수 숫자가 가장 작음)
        best_champ = mean_ranks.loc[mean_ranks['ranked'].idxmin()]
        # 가장 나쁜 챔피언 (등수 숫자가 가장 큼)
        worst_champ = mean_ranks.loc[mean_ranks['ranked'].idxmax()]
        
        results.append({
            'cost': cost,
            'p_value': p_value,
            'best_champion': best_champ['champion_name'],
            'best_mean_rank': best_champ['ranked'],
            'worst_champion': worst_champ['champion_name'],
            'worst_mean_rank': worst_champ['ranked']
        })

result_df = pd.DataFrame(results)
display(result_df)
```

### 16번: 같은 챔피언이라도 성급(star)에 따라 등수가 유의미하게 다르다.

```python
import pandas as pd
from scipy.stats import f_oneway

df_champ = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_champion_1.csv')
results = []

for champ in df_champ['champion_name'].dropna().unique():
    champ_data = df_champ[df_champ['champion_name'] == champ]
    
    star_groups = []
    mean_ranks = {}
    
    for star in sorted(champ_data['star'].dropna().unique()):
        ranks = champ_data[champ_data['star'] == star]['ranked']
        
        if len(ranks) > 5:
            star_groups.append(ranks.values)
            mean_ranks[f'star_{int(star)}_mean_rank'] = ranks.mean()
            
    if len(star_groups) > 1:
        stat, p_value = f_oneway(*star_groups)
        
        res_dict = {
            'champion': champ,
            'p_value': p_value
        }
        res_dict.update(mean_ranks)
        results.append(res_dict)

result_df = pd.DataFrame(results).sort_values('p_value')
display(result_df)
```

### 20번: 소수 정예 고단계 시너지(깊게) vs 다수 저단계 시너지(넓게)의 승률/순방률 비교

```python
import pandas as pd
from scipy.stats import ttest_ind
import ast

df_match = pd.read_csv('게임단위_게임데이터_상위랭커보존-2차 전처리 완료(match).csv')

def count_syn(x):
    if pd.isna(x):
        return 0
    if isinstance(x, str):
        if x.startswith('['):
            try:
                return len(ast.literal_eval(x))
            except:
                return len(x.split(','))
        return len(x.split(','))
    return 1

df_match['syn_count'] = df_match['combination'].apply(count_syn)

median_syn = df_match['syn_count'].median()

deep_deck = df_match[df_match['syn_count'] < median_syn]
wide_deck = df_match[df_match['syn_count'] > median_syn]

stat, p_value = ttest_ind(deep_deck['ranked'], wide_deck['ranked'], equal_var=False)

print(deep_deck['ranked'].mean())
print(wide_deck['ranked'].mean())

print(deep_deck['top4_flag'].mean())
print(wide_deck['top4_flag'].mean())

print(p_value)
```

다수 혼합(넓은 시너지) 덱의 평균 등수

### 21번: 같은 조합(시너지)이라도 어떤 챔피언을 메인 딜러/탱커로 쓰느냐에 따라 승률이 다르다

```python
import pandas as pd
from scipy.stats import f_oneway
import ast

df_match = pd.read_csv('게임단위_게임데이터_상위랭커보존-2차 전처리 완료(match).csv')
df_champ = pd.read_csv('게임단위_게임데이터_상위랭커보존-explode_champion_1.csv')

top_comp = df_match['combination'].value_counts().idxmax()
target_users = df_match[df_match['combination'] == top_comp][['gameid', 'user_id', 'ranked']]

df_target_champ = pd.merge(df_champ, target_users, on=['gameid', 'user_id', 'ranked'])

def count_items(x):
    if pd.isna(x):
        return 0
    if isinstance(x, str):
        x = x.strip()
        if x == '[]' or x == '':
            return 0
        if x.startswith('['):
            try:
                return len(ast.literal_eval(x))
            except:
                return len(x.split(','))
        return len(x.split(','))
    if isinstance(x, list):
        return len(x)
    return 1

df_target_champ['item_count'] = df_target_champ['items'].apply(count_items)

main_carries = df_target_champ[df_target_champ['item_count'] >= 2]

results = []
groups = []

for champ in main_carries['champion_name'].unique():
    ranks = main_carries[main_carries['champion_name'] == champ]['ranked']
    if len(ranks) > 10:
        groups.append(ranks.values)
        results.append({
            'main_carry': champ,
            'player_count': len(ranks),
            'mean_rank': ranks.mean()
        })

if len(groups) > 1:
    stat, p_value = f_oneway(*groups)
else:
    p_value = None

print(top_comp)
print(p_value)

result_df = pd.DataFrame(results).sort_values('mean_rank')
display(result_df)
```

제일 먼저 출력되는 문자열이 이번 분석의 타겟이 된 가장 유행하는 1티어 덱의 이름(조합)!

p_value가 0.05 미만이면 같은 덱을 하더라도 누구한테 템을 주느냐에 따라 승패가 완전히 갈린다는 뜻.

출력된 표에서 mean_rank가 가장 낮은(1등에 가까운) 챔피언이 그 덱의 진정한 핵심 캐리 기물이고, 반대로 mean_rank가 높은 챔피언은 템을 주면 안 되는 함정 기물이라는 걸 알 수 있다! 아마도...
